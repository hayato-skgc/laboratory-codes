```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# グラフに利用する値を定義
x = np.array([10.4, 13.7, 14.6, 16.9, 18.2, 20.0, 22.9, 25.0, 29.3])
y = np.array([19329, 7743, 6223, 4254, 3472, 2690, 1882, 1489, 925])
```

```python
# 通常のx-yグラフ
plt.scatter(x, y, color = 'black', marker='s')
plt.xlabel('Distant migrated (mm)')
plt.ylabel('Products (bp)')
plt.grid()
```

```python
# logy-xグラフ
log_y = np.log10(y)
linear = np.polyfit(x, log_y, 1)
linear_y = linear[0] * x + linear[1]

plt.yscale('log')
plt.scatter(x, y, color = 'black', marker='s')
plt.plot(x, 10**linear_y, color='red')
plt.xlabel('Distant migrated (mm)')
plt.ylabel('Products (log bp)')
plt.grid()

print('傾き: ' + str(linear[0]))
print('logy切片: ' + str(linear[1]))
```

```python
# 任意の移動距離から塩基対を求める
def get_bp(distant):
  log_bp = linear[0] * distant + linear[1]
  return 10**log_bp

print(get_bp(24.8))
```
